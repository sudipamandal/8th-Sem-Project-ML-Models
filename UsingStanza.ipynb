```python
import stanza

# Download an English model that supports sentiment analysis
stanza.download('en')

# Load the English model with sentiment analysis
nlp = stanza.Pipeline(lang='en', processors='tokenize,sentiment')

# Function to analyze sentiment and print negative or neutral sentences
def analyze_sentiment_and_print(paragraph):
    doc = nlp(paragraph)

    # Labels for sentiment scores
    sentiment_labels = {0: 'negative', 1: 'neutral', 2: 'positive'}

    # Iterate through each sentence in the paragraph
    for sentence in doc.sentences:
        # Check if the sentiment is negative or neutral
        if sentence.sentiment in [0, 1]:
            # Print the sentence and its sentiment label
            print(f"Sentence: {sentence.text}, Sentiment: {sentiment_labels[sentence.sentiment]}")
```
